The deliveries.csv file contains ball-by-ball data of cricket matches, with 150,460 entries and 21 columns. Here's a summary of its contents:

Purpose:
This dataset provides detailed information about every ball bowled in a cricket match, making it useful for:


Performance analysis of players (batsmen, bowlers, fielders).

Team strategy evaluation.

Statistical insights like runs scored, wickets taken, and extras conceded.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data= pd.read_csv("/content/deliveries.csv.zip")
data
```

```python
runs =data.groupby('batsman')
runs
```

```python
runs.get_group('RG Sharma')
```

A line plot is a type of data visualization used to display information as a series of data points connected by straight line segments. It is particularly effective for showing trends, changes, or patterns in data over a continuous range, such as time or sequences.

graph tells us:
Early Overs (1-6): Runs are usually moderate because batsmen are getting settled.
Middle Overs (7-15): Runs may dip as teams play more cautiously.
Death Overs (16-20): Runs typically go up because teams try to score as much as possible before the match ends.

```python
runs_per_over = data.groupby('over')['total_runs'].sum()
plt.figure(figsize=(10, 6))
sns.lineplot(x=runs_per_over.index, y=runs_per_over.values, marker="o")
plt.title("Total Runs Scored Per Over", fontsize=16)
plt.xlabel("Over", fontsize=12)
plt.ylabel("Total Runs", fontsize=12)
plt.xticks(range(1, 21))
plt.show()
```

A horizontal bar plot is a data visualization where categories are displayed along the vertical axis, and the corresponding values are represented by horizontal bars extending along the horizontal axis. This type of plot is often used when category names are long or when comparing a relatively small number of categories.

The bars represent the total number of balls each bowler has bowled.
The longer the bar, the more deliveries that bowler has bowled.
The graph highlights the most active bowlers in terms of participation across matches.

```python
top_bowlers = data['bowler'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_bowlers.values, y=top_bowlers.index, palette="viridis")
plt.title("Top 10 Bowlers by Ball Deliveries", fontsize=16)
plt.xlabel("Balls Delivered", fontsize=12)
plt.ylabel("Bowler", fontsize=12)
plt.show()
```

Each bar represents the total runs scored by a team across all their matches.
The longer the bar, the higher the total runs that team has scored.
Teams are ordered from the highest to the lowest scorers.

```python
team_performance = data.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 8))
sns.barplot(x=team_performance.values, y=team_performance.index, palette="coolwarm")
plt.title("Total Runs Scored by Each Team", fontsize=16)
plt.xlabel("Total Runs", fontsize=12)
plt.ylabel("Team", fontsize=12)
plt.show()
```

Heat Map is used to find out the correlation between different features in the dataset. High positive or negative value shows that the features have high correlation.This helps us to select the parmeters for machine learning.

You can compare which overs in each inning are high-scoring or low-scoring.
Usually, the death overs (16-20) show higher scores due to aggressive batting.
Early overs (1-6) might show lower scores as batsmen play cautiously.

```python
heatmap_data = data.pivot_table(
    index="over",
    columns="inning",
    values="total_runs",
    aggfunc="sum"
)

plt.figure(figsize=(10, 6))
sns.heatmap(
    heatmap_data,
    annot=True,
    fmt=".0f",
    cmap="YlGnBu",
    cbar_kws={'label': 'Total Runs'}
)
plt.title("Total Runs Scored Per Over by Innings", fontsize=16)
plt.xlabel("Inning", fontsize=12)
plt.ylabel("Over", fontsize=12)
plt.show()
```
